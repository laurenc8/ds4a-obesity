```python
x = df[cols[:-2]]/100
y = df[cols[-2]]/100
x = sm.add_constant(x) # adding a constant
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```

```python
print(model.params)
print("R^2:", model.rsquared)
```

# Cleaned EDA Code

```python
df_2021 = pd.read_csv('2021.csv')
df = df_2021.dropna(subset=['_BMI5', '_BMI5CAT'])[df_2021['_AGEG5YR'] != 14] # drop certain rows
cols = ['FTJUDA2_', 'FRUTDA2_', 'GRENDA1_', 'FRNCHDA_', 'POTADA1_', 'VEGEDA2_', '_BMI5', '_BMI5CAT'] # edit this list to include all the relevant columns
df = df[cols]
df['_BMI5'] /= 100 # divide by 100 (bc 2 implied decimal places)

# CLEANING DIET COLUMNS
for col in ['FTJUDA2_', 'FRUTDA2_', 'GRENDA1_', 'FRNCHDA_', 'POTADA1_', 'VEGEDA2_']:
    df[col] = df[col].replace(to_replace=9900) # replace 9900 with None
    df[col] = df[col].fillna(df[col].mean())/100 # impute None values with mean, divide by 100 (bc 2 implied decimal places)
```
